```python
from Bio.SeqIO.FastaIO import SimpleFastaParser
with open(SA) as fasta_file:  # Will close handle cleanly
    identifiers = []
    sequences = []
    for title, sequence in SimpleFastaParser(fasta_file):
        identifiers.append(title.split(None, 1)[0])  # First word is ID
        sequences.append(sequence)
```

```python
df_oa = pd.DataFrame([identifiers, sequences]).T
df_oa.columns = ['oligo','seq_25mer']
```

```python
transtable = str.maketrans("ACTG","TGAC")
```

```python
df4['seq_25mer_rv'] = df4.seq_25mer_fw.apply(lambda x: x.translate(transtable)[::-1])
```

```python
df4['seq_15mer_rv'] = df4.seq_15mer_fw.apply(lambda x: x.translate(transtable)[::-1] if x == x else np.nan)
```

```python
df4[['id_25mer','seq_25mer_fw','seq_25mer_rv',
     'id_15mer_1','id_15mer_2','seq_15mer_fw','seq_15mer_rv',
     'oligo']].to_csv(OP, index=False)
```
