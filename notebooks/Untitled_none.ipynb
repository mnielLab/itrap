```python
PREDICTIONS = 'nettcr_predictions.cropped.csv'
```

```python
pred = pd.read_csv(PREDICTIONS)
```

```python
pred
```

```python
n = pd.merge(pred, df[['peptide','cdr3_TRA','cdr3_TRB','peptide_HLA_lst','label']],
             left_on=['peptide','CDR3a','CDR3b'],
             right_on=['peptide','cdr3_TRA','cdr3_TRB'],
             how='outer')
n.drop_duplicates(subset=['peptide','CDR3a','CDR3b','peptide_HLA_lst'], inplace=True)
```

```python
groups = n.groupby(['CDR3a','CDR3b'])
```

```python
for cdr3, grp in groups:
    grp.sort_values(by=['percentile_rank','prediction'], ascending=[True,False], inplace=True)
    print(grp.index)
    n.loc[grp.index, 'pred'] = [True, False, False]
    n.loc[grp.index, 'labl'] = ~grp.peptide_HLA_lst.isna() #~top_hit.isna().any().any()
    n.loc[grp.index, 'tied'] = grp.duplicated(subset='percentile_rank', keep=False) #~grp[dup].isna().all().any()
```
