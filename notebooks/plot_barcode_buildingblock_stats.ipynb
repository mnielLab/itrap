```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
from glob import iglob
```

```python
plt.style.use('ggplot')
```

# Args

```python
EXP = "exp3"
```

# Plotting directory

```python
FIG_DIR = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/plots/" + EXP + "/"
```

# Load data

```python
BARCODE_BUILDINGBLOCK_STATS_DIRECTORY = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/data/" + EXP + "_MHC/barcode_buildingblock_stats/IONTORRENT.R1.gems.no_umi.no_adapters/"
```

```python
data_dict = dict()

for filename in iglob(BARCODE_BUILDINGBLOCK_STATS_DIRECTORY + '*.lst', recursive=False):
    if os.path.isfile(filename):
        filter_key = filename.split("/")[-1].split(".")[0].split("_", 2)[-1]
        data_dict[filter_key] = np.loadtxt(filename)
        print(filter_key)
```

```python
fig = plt.figure(figsize=(10,12))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i, key in enumerate(data_dict.keys()):
    plt.subplot(5, 3, i+1)
    n, bins, patches = plt.hist(data_dict[key], bins=np.arange(1, 300, 1))
    patches[np.where(bins==149)[0][0]].set_fc('black')
    plt.title(key)
    if i+1 in [1, 4, 7, 10, 13]:
        plt.ylabel("Read counts")
    if i+1 in [12, 13, 14]:
        plt.xlabel("Read length")
fig.suptitle("Experiment 3", fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.92)
plt.savefig(FIG_DIR + "read_length_distributions.pdf")
```
