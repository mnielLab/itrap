```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Input

```python
POSITIVE = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/data/net1_TCR/augmented/tcr.clean.augmented.csv"
NEGATIVE = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/data/net2_TCR/augmented/tcr.clean.augmented.csv"
```

```python
pos_df = pd.read_csv(POSITIVE)
neg_df = pd.read_csv(NEGATIVE)
```

```python
pos_df['target'] = 1
neg_df['target'] = 0
```

```python
x = pos_df[pos_df['exclude_ambiguous_and_single-chain_TCRs'] == True].groupby('clonotype').gem.apply(list).apply(len)

plt.figure(figsize=(15,5))
plt.axhline(1, c='r')
plt.plot(x.sort_values(ascending=False))
plt.xticks(rotation=90, size=8)

plt.ylabel('GEM count')
plt.title('GEMs per clonotype')
plt.show()
```

```python
x.sort_values().tail(60)
```

```python
x = pos_df[pos_df['exclude_ambiguous_and_single-chain_TCRs'] == True].groupby('ct').gem.apply(list).apply(len)

plt.figure(figsize=(15,5))
plt.plot(x.sort_values(ascending=False))
plt.xticks(rotation=90, size=8)

plt.ylabel('GEM count')
plt.title('GEMs per clonotype')
plt.show()
```

```python
x = pos_df[pos_df['exclude_ambiguous_and_single-chain_TCRs'] == True].groupby(['cdr3_TRA','cdr3_TRB']).gem.apply(list).apply(len)
x.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(15,5))
plt.scatter([' '.join(map(str,i)) for i in x.index.tolist()], x.values)
plt.xticks(rotation=90, size=8)

plt.ylabel('GEM count')
plt.title('GEMs per clonotype')
plt.show()
```

```python
x = pos_df[pos_df['exclude_ambiguous_and_single-chain_TCRs'] == True].groupby(['cdr3_TRA','cdr3_TRB']).num_clonotype.unique().explode()
x.dropna(inplace=True)
x.sort_values(inplace=True)

plt.figure(figsize=(15,15))
plt.scatter([' '.join(map(str,i)) for i in x.index.tolist()], x.values)
plt.xticks(rotation=90, size=8)

plt.ylabel('Clonotypes')
plt.title('Clonotypes per CDR3 combination')
plt.show()
```

```python
x = pos_df[pos_df['exclude_ambiguous_and_single-chain_TCRs'] == True].groupby(['cdr3_TRA','cdr3_TRB']).ct.unique().explode()
x.sort_values(inplace=True)

plt.figure(figsize=(15,15))
plt.scatter([' '.join(map(str,i)) for i in x.index.tolist()], x.values)
plt.xticks(rotation=90, size=8)

plt.ylabel('Clonotypes')
plt.title('Clonotypes per CDR3 combination')
plt.show()
```
