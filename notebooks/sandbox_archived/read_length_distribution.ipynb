```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
```

```python
plt.style.use('ggplot')
```

# Args

```python
EXP = "exp6"
PRJ = 'CTPG5' #'C8H64'
```

# Input

```python
R1 = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/data/" + EXP + "_TCR/processed/mkfastq_out/" + PRJ + "/TCR/R1_read_lengths.txt"
R2 = "/Volumes/tuba/herpov/tcr-pmhc-sc-project/data/" + EXP + "_TCR/processed/mkfastq_out/" + PRJ + "/TCR/R2_read_lengths.txt"
```

# Output

```python
directory = 'stats/' + EXP + '/'
```

# Load

```python
r1 = pd.read_csv(R1, sep=' ', names=['read_length', 'frequency'])
r2 = pd.read_csv(R2, sep=' ', names=['read_length', 'frequency'])
```

# Main

```python
plt.figure(figsize=(15,7))
plt.subplot(2,1,1)
x = r1.read_length
y = r1.frequency
plt.bar(x,y)
plt.xlim((min(x)-0.5, max(x)+0.5))
plt.xlabel('R1 read lengths')
plt.ylabel('Frequency')
plt.title('Distribution of R1 read lengths')

plt.subplot(2,1,2)
x = r2.read_length
y = r2.frequency
plt.bar(x,y)
plt.xlim((min(x)-0.5, max(x)+0.5))
plt.xlabel('R2 read lengths')
plt.ylabel('Frequency')
plt.title('Distribution of R2 read lengths')

plt.subplots_adjust(hspace = 0.5)
plt.suptitle('Raw data')
plt.savefig(directory + "raw_read_length_distr.pdf")
plt.show()
```
