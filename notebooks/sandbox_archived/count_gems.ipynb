```python
for k,v in tcr.items():
    if ('tra' in k) or ('trb' in k):
        break
    for q,w in mhc.items():
        if ('mhc' in q) or ('cdx' in q):
            break
        
        a = set(v)
        b = set(w)
        
        ab = a.intersection(b)
        a_b = a - ab
        b_a = b - ab
        
        venn2(subsets = (len(a_b), len(b_a), len(ab)), set_labels = ('TCR', 'Barcode'))
        plt.title('%s (%d TCRs)\n%s (%d BCs)'%(k, len(v), q, len(w)))
        plt.show()
        
    for q,w in cat.items():
        print('cat')
        print(q)
        
        a = set(v)
        b = set(w)
        
        ab = a.intersection(b)
        a_b = a - ab
        b_a = b - ab
        
        venn2(subsets = (len(a_b), len(b_a), len(ab)), set_labels = ('TCR', 'Barcode'))
        plt.title('%s (%d TCRs)\n%s (%d BCs)'%(k, len(v), q, len(w)))
        plt.show()
```
