## Plotting functions for analysing non-intersecting reads

```python
def kma_diagnostics(x, kma_set, directory, title):
    plt.subplot(121)
    plt.hist(kma_set[kma_set.read_barcode.isin(x)].uncertainty, bins=np.arange(kma_set[kma_set.read_barcode.isin(x)].uncertainty.max()+2))
    plt.title("Uncertainty")
    plt.ylabel("Frequency")
    plt.subplot(122)
    plt.hist(kma_set[kma_set.read_barcode.isin(x)].score, bins=np.arange(kma_set[kma_set.read_barcode.isin(x)].score.max()+2))
    plt.title("Score")
    plt.savefig(directory + "_".join(title.split()) + ".pdf", bbox_inches='tight')
    plt.show()
```

```python
def blast_diagnostics(x, blast_set, directory, title):
    ann_df = blast_set[blast_set.read_barcode.isin(x)]
    
    if ann_df.empty:
        print("The blast set does not contain any of the queried entries")
        return 

    # Plot
    cc = plt.rcParams['axes.prop_cycle'].by_key()['color'] * 2
    data_categories = {'expected_templates':{'TSO':(1, 13, cc[3]),
                                             'Primer_B':(14, 19, cc[0]),
                                             'N6':(33, 6, cc[4]),
                                             'Oligo_B':(39, 25, cc[2]),
                                             'Anneal':(64, 25, cc[1]),
                                             'Oligo_A':(89, 25, cc[2]),
                                             'N6-':(114, 6, cc[4]),
                                             'Primer_A':(120, 18, cc[0]),
                                             'ID':(138, 8, cc[5])}}

    # Set up the axes with gridspec
    fig = plt.figure(figsize=(14, 6))
    grid = plt.GridSpec(2, 1, hspace=0.25, height_ratios=[10,1])
    # Main
    main_ax = plt.subplot(grid[0, 0])

    #for i in range(ann_df.match.max() + 1):
    #    main = plt.plot([ann_df[ann_df.match == i].t_alignment_start.to_list(), ann_df[ann_df.match == i].t_alignment_end.to_list()],
    #                    [ann_df[ann_df.match == i].q_alignment_start.to_list(),ann_df[ann_df.match == i].q_alignment_end.to_list()],
    #                    '.-', c=cc[i], label="%i block matches" %i)
    for i in range(2):
        main = plt.plot([ann_df[ann_df.blast_pass == i].t_alignment_start.to_list(), ann_df[ann_df.blast_pass == i].t_alignment_end.to_list()],
                        [ann_df[ann_df.blast_pass == i].q_alignment_start.to_list(),ann_df[ann_df.blast_pass == i].q_alignment_end.to_list()],
                        '.-', c=cc[i], label="%i block matches" %i)
    main = plt.xlim(-3,153) # Variable!
    main = plt.xlabel("Alignment positions in template")
    main = plt.ylabel("Alignment positions in query")
    main = plt.title("Alignment positions")
    # Subs
    subs_ax = plt.subplot(grid[1, 0], yticklabels=[], sharex=main_ax)
    # Add template design
    for k, v in data_categories['expected_templates'].items():
        # Create a Rectangle patch
        rect = patches.Rectangle((v[0],0),v[1],1, fill=True, color=v[2]) #linewidth=1,edgecolor='r',facecolor='none'
        subs_ax.add_patch(rect)
        # Add text
        block = " ".join(k.split("_")).split("-")[0]
        txt = subs_ax.text(v[1]/2.0+v[0], 0.5, block, horizontalalignment='center', verticalalignment='center', color='white',
                           size=15, weight='heavy')
        txt.set_path_effects([PathEffects.withStroke(linewidth=0.5, foreground='black')])
        
        plt.savefig(directory + "_".join(title.split()) + ".pdf", bbox_inches='tight')
```

```python
def and_diagnostics(x, and_set, directory, title):
    fig = plt.figure(figsize=(12,6))
    plt.subplot(121)
    plt.plot(and_set[and_set.read_barcode.isin(x) & (and_set.align_object == "A_OLIGO")].align_score1.values,
             and_set[and_set.read_barcode.isin(x) & (and_set.align_object == "A_OLIGO")].align_score2.values,
             '.', color='#9e0142')
    plt.title("Alignemnt scores of oligo A")
    plt.xlabel("Best alignment score")
    plt.ylabel("Second best alignment score")
    plt.xlim(min(and_set[and_set.align_object == "A_OLIGO"].align_score1.min(), and_set[and_set.align_object == "B_OLIGO"].align_score1.min()),
             max(and_set[and_set.align_object == "A_OLIGO"].align_score1.max(), and_set[and_set.align_object == "B_OLIGO"].align_score1.max()) + 5)
    plt.ylim(min(and_set[and_set.align_object == "A_OLIGO"].align_score2.min(), and_set[and_set.align_object == "B_OLIGO"].align_score2.min()),
             max(and_set[and_set.align_object == "A_OLIGO"].align_score2.max(), and_set[and_set.align_object == "B_OLIGO"].align_score2.max()) + 5)
    
    plt.subplot(122)
    plt.plot(and_set[and_set.read_barcode.isin(x) & (and_set.align_object == "B_OLIGO")].align_score1.values,
             and_set[and_set.read_barcode.isin(x) & (and_set.align_object == "B_OLIGO")].align_score2.values,
             '.', color='#9e0142')
    plt.title("Alignemnt scores of oligo B")
    plt.xlabel("Best alignment score")
    plt.ylabel("Second best alignment score")
    plt.xlim(min(and_set[and_set.align_object == "A_OLIGO"].align_score1.min(), and_set[and_set.align_object == "B_OLIGO"].align_score1.min()),
             max(and_set[and_set.align_object == "A_OLIGO"].align_score1.max(), and_set[and_set.align_object == "B_OLIGO"].align_score1.max()) + 5)
    plt.ylim(min(and_set[and_set.align_object == "A_OLIGO"].align_score2.min(), and_set[and_set.align_object == "B_OLIGO"].align_score2.min()),
             max(and_set[and_set.align_object == "A_OLIGO"].align_score2.max(), and_set[and_set.align_object == "B_OLIGO"].align_score2.max()) + 5)
    
    plt.savefig(directory + "_".join(title.split()) + ".pdf", bbox_inches='tight')
    plt.show()
```

```python
def score_diagnostics(x, x_df, directory, title):
    # plot alignemnt scores from BLAST against KMA (color code by alignment length?)
    # x : set
    # x_df : sub set of merged dataframes
    
    x_df.dropna(subset=['bit_score', 'score', 'alignment_length'], inplace=True)
    colors = x_df.alignment_length.values / 150
    
    sm = plt.scatter(x_df.bit_score, x_df.score, marker='.', c=colors, cmap='viridis_r')
    plt.colorbar(sm)
    plt.xlabel("BLAST scores")
    plt.ylabel("KMA scores")
    plt.title("Compare scores (color coded by alignment length)")
    plt.savefig(directory + "_".join(title.split()) + ".pdf", bbox_inches='tight')
    plt.show()
```

## Intersections between read ID & barcode while passing thresholds

```python
# Get intersections of read_id+barcode
#kma_set['read_barcode'] = kma_set['read_id'] + kma_set['kma_barcode']
#eval_set['read_barcode'] = eval_set['read_id'] + eval_set['and_barcode']
#blast_set['read_barcode'] = blast_set['query_id'] + blast_set['blast_barcode']

uncertainty_threshold = 1

# Assumes that inputs are sets
a, b, c = set(eval_set[eval_set.and_pass == 1].read_barcode), set(blast_set[blast_set.blast_pass == 1].read_barcode), set(kma_set[kma_set.uncertainty <= uncertainty_threshold].read_barcode)
#a, b, c = set(eval_set[eval_set.and_pass == 1].read_barcode), set(blast_set.read_barcode), set(kma_set[kma_set.uncertainty <= uncertainty_threshold].read_barcode)


ab = a.intersection(b)
ac = a.intersection(c)
bc = b.intersection(c)

abc = ab.intersection(c)
ab_ = ab - abc
ac_ = ac - abc
bc_ = bc - abc
a__ = a - ab_ - ac_ - abc
b__ = b - ab_ - bc_ - abc
c__ = c - ac_ - bc_ - abc

from matplotlib_venn import venn3

title = "Barcode annotation comparison"
# Make the diagram
venn3(subsets = (len(a__), len(b__), len(ab_), len(c__), len(ac_), len(bc_), len(abc)), set_labels = ('Andrea', 'BLAST', 'KMA'))
plt.title(title)
plt.savefig(FIG_DIR + "_".join(title.split(" ")) + ".pdf", bbox_inches='tight')
plt.show()
```

```python
compare_2d_ke = pd.DataFrame(index=sorted(kma_set.uncertainty.unique()), columns=['uncertainty_pass','barcode_U','Apass_U_Kpass','Barcode_U_A_U_K'])
compare_2d_kb = pd.DataFrame(index=sorted(kma_set.uncertainty.unique()), columns=['uncertainty_pass','barcode_U','Bpass_U_Kpass','Barcode_U_B_U_K'])
compare_3d = pd.DataFrame(index=sorted(kma_set.uncertainty.unique()), columns=['uncertainty_pass','barcode_U','Apass_U_Kpass_U_Bpass','Barcode_U_A_U_K_U_B'])
```

```python
for threshold in sorted(kma_set.uncertainty.unique()):
    df['uncertainty_pass'] = np.where(df.uncertainty > threshold, 0, 1)
    df['barcode_union'] = np.where(df.kma_barcode == df.and_barcode, 1, 0)
    df['union_pass'] = np.where(df.and_pass == df.uncertainty_pass, 1, 0)
    
    compare_2d_ke.loc[threshold, 'uncertainty_pass'] = df.uncertainty_pass.sum()
    compare_2d_ke.loc[threshold, 'barcode_U'] = df[df.uncertainty_pass == 1].barcode_union.sum()
    compare_2d_ke.loc[threshold, 'Apass_U_Kpass'] = df[df.uncertainty_pass == 1].and_pass.sum()
    compare_2d_ke.loc[threshold, 'Barcode_U_A_U_K'] = df[(df.uncertainty_pass ==1) & (df.and_pass == 1)].barcode_union.sum()

    df2['uncertainty_pass'] = np.where(df2.uncertainty > threshold, 0, 1)
    df2['barcode_union'] = np.where(df2.kma_barcode == df2.blast_barcode, 1, 0)
    df2['union_pass'] = np.where(df2.blast_pass == df2.uncertainty_pass, 1, 0)
    
    compare_2d_kb.loc[threshold, 'uncertainty_pass'] = df2.uncertainty_pass.sum()
    compare_2d_kb.loc[threshold, 'barcode_U'] = df2[df2.uncertainty_pass == 1].barcode_union.sum()
    compare_2d_kb.loc[threshold, 'Bpass_U_Kpass'] = df2[df2.uncertainty_pass == 1].blast_pass.sum()
    compare_2d_kb.loc[threshold, 'Barcode_U_B_U_K'] = df2[(df2.uncertainty_pass ==1) & (df2.blast_pass == 1)].barcode_union.sum()

    df3['uncertainty_pass'] = np.where(df3.uncertainty > threshold, 0, 1)
    df3['barcode_union'] = np.where((df3.kma_barcode == df3.and_barcode) & (df3.kma_barcode == df3.blast_barcode), 1, 0)
    df3['union_pass'] = np.where((df3.uncertainty_pass == df3.and_pass) & (df3.uncertainty_pass == df3.blast_pass), 1, 0)

    compare_3d.loc[threshold, 'uncertainty_pass'] = df3.uncertainty_pass.sum()
    compare_3d.loc[threshold, 'barcode_U'] = df3[df3.uncertainty_pass == 1].barcode_union.sum()
    compare_3d.loc[threshold, 'Apass_U_Kpass_U_Bpass'] = df3[df3.uncertainty_pass == 1].and_pass.sum()
    compare_3d.loc[threshold, 'Barcode_U_A_U_K_U_B'] = df3[(df3.uncertainty_pass ==1) & (df3.and_pass == 1)].barcode_union.sum()
```
